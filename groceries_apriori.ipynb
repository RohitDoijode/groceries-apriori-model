```python
#Importing Pandas Library
import pandas as pd
```

```python
# Reading the dataset and appending it to the list
dataset = pd.read_csv('groceries.csv', header = None)
transactions = []
for i in range(0, 9835):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 32)])
```

```python
#Training apriori model on the dataset
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.002, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
results = list(rules)
```

```python
#Displaying  Results
results
```

```python
#Displaying results in an pandas dataframe
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
#sorting results according to the decreasing order of Lift
resultsinDataFrame.nlargest(n = 20, columns = 'Lift')
```
